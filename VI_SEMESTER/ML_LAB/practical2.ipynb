```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Diwali Sales Data.csv",encoding= 'unicode_escape')
df
```

```python
df.drop(columns=['unnamed1'], inplace=True)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='Product_Category', y='Amount', palette='viridis')
plt.title('Sales Amount by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Sales Amount')
plt.xticks(rotation=45, ha='right')  # Rotate the x-axis labels for better readability
plt.tight_layout()  # Adjust layout for better visualization
plt.show()
```

```python
# Step 4: Gender-wise Sales Distribution
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='Gender', y='Amount', palette='coolwarm')
plt.title('Sales Amount Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('Sales Amount')
plt.show()
```

```python
# Step 5: Age Group vs. Sales Amount (using barplot)
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='Age Group', y='Amount', hue='Gender', palette='Set2')
plt.title('Sales Amount by Age Group and Gender')
plt.xlabel('Age Group')
plt.ylabel('Sales Amount')
plt.show()
```

```python
# Step 6: Orders Distribution across Different Occupations
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Occupation', hue='Gender', palette='Set1')
plt.title('Number of Orders by Occupation and Gender')
plt.xlabel('Occupation')
plt.ylabel('Number of Orders')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
numerical_df = df.select_dtypes(include=['float64', 'int64'])  # Select only numerical columns
correlation_matrix = numerical_df.corr()  # Find correlations between numerical columns
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```
